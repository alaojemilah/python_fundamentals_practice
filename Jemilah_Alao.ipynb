Write a program that counts the total number of vowels and consonants in a given sentence, ignoring spaces, special characters and numbers. Example, if given the sentence "Loops are Fun!" Vowels: 5 Consonants: 6

```python
sentence = 'I love Python'
vowels = 0
consonants = 0

for letter in sentence.lower():
    if letter in "aeiou":
        vowels += 1
    elif letter.isalpha():
        consonants += 1
print("vowels:", vowels)
print("Consonants:", consonants)
```

## Question 4
Write a program to use a loop to find the factorial of a given number (e.g. 5!). The factorial of N is the product of all integers from 1 to N. Example, if given 5, the factorial is 120 Note: factorial does not exist for negative numbers, the factorial of 0 is 1.



```python
number = 6

factorial = 1

if number < 0:
    print("Factorial does not exist for negative numbers")

elif number == 0:
    print("Factorial of 0 is 1")

else:
    for i in range(1, number + 1):
        factorial *= i

    print("Factorial =", factorial)
```

## Question 1
Write a program that sorts an array of planet names in ascending order based on their distance from the Sun.
Planet Order (from closest to farthest): Mercury, Venus, Earth, Mars, Jupiter, Saturn, Uranus, Neptune.
Constraints:
The input array may contain planet names in any order and in mixed case (e.g., "mars", "Venus", "EARTH").
The output should be sorted correctly and have all names in title case (e.g., "Earth").
Example Input: ["Neptune", "earth", "Mars", "Venus"
"jupiter"]
Expected Output: ["Venus", "Earth", "Mars", "Jupiter"
"Neptune"]

```python

planets_order = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"]
planets = ["earth", "Nepturn", "Uranus", "Jupiter", "mars", "VENUS"]
sorted_planets = []

for i in range(len(planets)):
    planets[i] = planets[i].title()
    
for planet in planets_order:
    for p in planets:
        if p == planet:
            sorted_planets.append(p)
print(sorted_planets)
```

## Question 2
An online store calculates shipping costs based on order value and delivery distance.
Orders over N100 get free shipping. For orders under N100, shipping is N5 if the distance is under 10 miles, and N10 if it is 10 miles or more. You need to process a batch of order totals and distances to find the total revenue including shipping. starter code:
Data: List of tuples (order_value, distance_miles)
orders = [(120, 5), (45, 12), (85, 3), (200, 15), (30, 8)]
total
_revenue = 0

```python
#List of tuples(order_value, distance_miles) 
orders = [(120, 5), (45, 12), (85, 3), (200, 15), (30, 8)] 
total_revenue = 0

for order in orders:
    order_value = order[0]
    distance = order[1]
    
    if order_value > 100:
        shipping_fee = 0
    elif distance < 10:
        shipping_fee = 5
    else:
        shipping_fee = 10
    total_revenue = total_revenue + order_value + shipping_fee
print("Total revenue: N" + str(total_revenue))
```

## Question 3
An industrial solar battery drains at night. It starts at
100% capacity. Every hour, it loses 8% of its current remaining capacity. If the battery capacity drops below 40%, a backup generator must switch on. Determine how many hours the battery lasts before the generator kicks in.

```python
capacity = 100
hours = 0

while capacity > 40:
    capacity = capacity - (capacity * 8 / 100)
    hours = hours +1
print(str(hours) + "hours")
```

Example: if given USD 5, it should print NGN6805.30.

```python
def convert_currency(dollar):
    """convert_currency converts dollar to naira using a fixed rate of 1,361.06"""
    value = dollar * 1361.06
    print(f"NGN{value:.2f}.") # 2f means format the number as a fixed point decimal i.e show exactly 2 digits after the decimal point
```

## Question 2
Calculate Discounted Price: A store gives a
5% discount on items bought. Write a function called calculate_discount that calculates the final price after a percentage discount. This function also takes two arguments.
Example: If I buy goods worth #2000 and I have 5% discount, my final price would be #1900

```python
def calculate_discount(price, discount_rate):
    """calculates the final price after a percentage discount"""
    discount = int(price - (price * discount_rate / 100))
    return discount
```

```python
print('#' + str(calculate_discount(2000, 5)))
```

## Question 3
Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
def find_numbers():
    """find_numbers between a range of 2000 and 3200(both included)"""
    result = []
    for num in range(2000, 3201):
        if num % 7 == 0 and num % 5 != 0:
            result.append(str(num))
    return ','.join(result)    #join takes every item on the result list and connects them with the seperator.
print(find_numbers())
```

```python
def fibonacci(n):
    #base case
    if n == 0:
        return 0
    elif n == 1:
        return 1
        #recursive case
    else:
        # print(n)
        return fibonacci(n-1) + fibonacci (n-2)
```
